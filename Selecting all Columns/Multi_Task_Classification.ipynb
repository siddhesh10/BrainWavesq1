```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
%matplotlib inline
```

```python
train = pd.read_csv('Updated_train.csv')
test= pd.read_csv('Updated_test.csv')
```

```python
X_train, y_train = train['new'].values, train['Complaint-Status'].values
X_test = test['new'].values
```

```python
y_train
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def tfidf_features(X_train, X_test):
    
    tfidf_vectorizer = TfidfVectorizer(sublinear_tf=True, min_df=5, norm='l2', encoding='latin-1', ngram_range=(1, 2))
    X_train = tfidf_vectorizer.fit_transform(X_train)
    X_test = tfidf_vectorizer.transform(X_test)
    return X_train, X_test, tfidf_vectorizer.vocabulary_
```

```python
X_train_tfidf, X_test_tfidf, tfidf_vocab = tfidf_features(X_train,X_test)
tfidf_reversed_vocab = {i:word for word,i in tfidf_vocab.items()}
```

```python
print('X_train_tfidf shape', X_train_tfidf.shape)
print('X_test_tfidf shape ', X_test_tfidf.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC

mnb = MultinomialNB().fit(X_train_tfidf, y_train)
lr = LogisticRegression().fit(X_train_tfidf, y_train)
rfc = RandomForestClassifier().fit(X_train_tfidf, y_train)
clf = LinearSVC().fit(X_train_tfidf, y_train)
```

```python
mnb_a= list(mnb.predict(X_test_tfidf))
lr_a= list((lr.predict(X_test_tfidf)))
rfc_a= list(rfc.predict(X_test_tfidf))
clf_a= list(clf.predict(X_test_tfidf))
```

```python
#Closed with non-monetary relief - 3
#Closed with explanation - 1
#Untimely response- 4
#Closed with monetary relief -2
#Closed-0
```

```python
converting_dictionary = {'Closed':0,'Closed with explanation':1,'Closed with monetary relief':2,'Closed with non-monetary relief':3,'Untimely response':4}
inv_map = {v: k for k, v in converting_dictionary.items()}
```

```python
for i in range(len(mnb_a)):
    mnb_a[i]=inv_map[mnb_a[i]]
```

```python
for i in range(len(lr_a)):
    lr_a[i]=inv_map[lr_a[i]]

for i in range(len(rfc_a)):
    rfc_a[i]=inv_map[rfc_a[i]]
    
for i in range(len(clf_a)):
    clf_a[i]=inv_map[clf_a[i]]
```

```python
ans_mnb= pd.DataFrame({'Complaint-Status':mnb_a})
ans_lr= pd.DataFrame({'Complaint-Status':lr_a})
ans_rfc= pd.DataFrame({'Complaint-Status':rfc_a})
ans_clf= pd.DataFrame({'Complaint-Status':clf_a})
```

```python
ans_mnb['Complaint-ID']=test['Complaint-ID']
ans_mnb=ans_mnb[['Complaint-ID','Complaint-Status']]

ans_lr['Complaint-ID']=test['Complaint-ID']
ans_lr=ans_lr[['Complaint-ID','Complaint-Status']]

ans_rfc['Complaint-ID']=test['Complaint-ID']
ans_rfc=ans_rfc[['Complaint-ID','Complaint-Status']]

ans_clf['Complaint-ID']=test['Complaint-ID']
ans_clf=ans_clf[['Complaint-ID','Complaint-Status']]
```
